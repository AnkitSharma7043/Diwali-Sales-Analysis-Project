```python
import matplotlib.pyplot as plt
plt.title=plt.suptitle # restore title in matplot
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers: # for loop use for labels
    ax.bar_label(bars)

plt.title('Male and Female')
plt.show()
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
df.columns
```

```python
ax=sns.countplot(x='Age Group',data=df,hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# total amount vs age group
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sales_age
```

 most of the buyers are of age group between 26-35yrs male/female

```python
## top 10 states by orders
sales_st=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
```

```python
sales_st
```

```python
sales_st=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,6)})
sns.barplot(data=sales_st,x='State',y='Orders')
```

```python
# total states by amount/sales
sales_st=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,6)})
sns.barplot(data=sales_st,x='State',y='Amount')                                                                                                           
```

most of the orders and total sales/amount are from u.p,maharashtra and karnataka
